```python
import os
import sys
import json
import itertools
from scipy import signal
import numpy
import matplotlib.pyplot as pyplot
```

```python
pyplot.rc("font", family="serif", size=10, weight="bold")
pyplot.rc("axes", labelweight="bold", titleweight="bold")
pyplot.rc("text", usetex=True)
figure =  pyplot.figure(figsize=(4,3), dpi=200)

numdatasets = [6, 12, 30]
nthreads1 = [27.888, 50.886, 121.906]
nthreads8 = [10.179, 16.983, 38.897]
nthreads12 = [10.723, 16.029, 40.007]

pyplot.plot(numdatasets, nthreads1, linestyle = '-', color='black', marker='s')
pyplot.plot(numdatasets, nthreads8, linestyle = '--', color='red', marker='d')
pyplot.plot(numdatasets, nthreads12, linestyle = '-.', color='blue', marker='o')

pyplot.ylim([0,150])
pyplot.yticks([0,50,100,150])
pyplot.ylabel(r"Time (s)", rotation=0, labelpad=5)
pyplot.xlabel(r"Datasets", labelpad=1)
pyplot.legend(["Threads = 1", "Threads = 8", "Threads = 12"])
pyplot.title("Computation of temporal mean")
pyplot.tight_layout()
pyplot.savefig("/Users/Akash/Desktop/one.png")
```
